```python
import pandas as pd
import numpy as np
```

```python
app_proj = pd.read_csv('UPDATED_approved_projects_with_demographics.csv')
app_zba = pd.read_csv('Approved_Permits_ZBA - zba_approved_permits.csv')
# parcel id column is titled parcel_id in both
```

```python
for i in range(len(app_proj['issued_date'])):
    string = app_proj.loc[i, 'issued_date']
    year = string[-8:-6]
    if '/' in year:
        year = string[-7:-5]
    app_proj.loc[i, 'issued_date'] = '20' + year
```

```python
app_proj['issued_date'].unique()
```

```python
for i in range(len(app_zba['issued_date'])):
    if type(app_zba.loc[i, 'issued_date']) != float:
        string = app_zba.loc[i, 'issued_date']
        year = string[0:4]
        app_zba.loc[i, 'issued_date'] = year
```

```python
app_zba['issued_date'].unique()
```

```python
adds = pd.DataFrame(columns = ['parcel_id', 'Owner.1', 'Owner.2', 'Address'])
adds['parcel_id'] = app_proj['parcel_id']
adds['Owner.1'] = app_proj['owner']
adds.head()
```

```python
pa_21 = pd.read_csv('Property Assesments/property-assessment-fy2021.csv')
pa_21 = pa_21.astype({"PID": str})
pa_21.head()
```

```python
for i in range(len(adds)):
    pid = adds.loc[i, 'parcel_id']
    filt = pa_21['PID'] == pid
    temp = pa_21[filt]
    temp = temp.reset_index(drop=False)
    if len(temp) == 0:
        pass
    else:
        adds.loc[i, 'Owner.2'] = temp.loc[0, 'OWNER']
        adds.loc[i, 'Address'] = str(temp.loc[0, 'MAIL_ADDRESS']) + ' ' + str(temp.loc[0, 'MAIL_CITY']) + \
                                 ' ' + str(temp.loc[0, 'MAIL_STATE']) + ' ' + str(temp.loc[0, 'MAIL_ZIPCODE'])
adds.head()     
```

```python
adds.to_csv('PID_Owner_Comparison_And_Address.csv')
```
