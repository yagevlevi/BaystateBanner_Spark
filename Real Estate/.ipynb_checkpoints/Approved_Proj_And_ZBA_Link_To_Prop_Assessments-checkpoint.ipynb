```python
import pandas as pd
import numpy as np
```

```python
app_proj = pd.read_csv('UPDATED_approved_projects_with_demographics.csv')
app_zba = pd.read_csv('Approved_Permits_ZBA - zba_approved_permits.csv')
# parcel id column is titled parcel_id in both
```

```python
for i in range(len(app_proj['issued_date'])):
    string = app_proj.loc[i, 'issued_date']
    year = string[-8:-6]
    if '/' in year:
        year = string[-7:-5]
    app_proj.loc[i, 'issued_date'] = '20' + year
```

```python
for i in range(len(app_zba['issued_date'])):
    if type(app_zba.loc[i, 'issued_date']) != float:
        string = app_zba.loc[i, 'issued_date']
        year = string[0:4]
        app_zba.loc[i, 'issued_date'] = year
```

```python
adds = pd.DataFrame(columns = ['parcel_id', 'Owner.1', 'Owner.2', 'Address', 'Use_Code'])
adds['parcel_id'] = app_proj['parcel_id']
adds['Owner.1'] = app_proj['owner']
adds.head()
```

```python
pa_21 = pd.read_csv('Property Assesments/property-assessment-fy2021.csv')
pa_21 = pa_21.astype({"PID": str})
pa_21.head()
```

```python
for i in range(len(adds)):
    pid = adds.loc[i, 'parcel_id']
    filt = pa_21['PID'] == pid
    temp = pa_21[filt]
    temp = temp.reset_index(drop=False)
    if len(temp) == 0:
        pass
    else:
        adds.loc[i, 'Owner.2'] = temp.loc[0, 'OWNER']
        adds.loc[i, 'Address'] = str(temp.loc[0, 'MAIL_ADDRESS']) + ' ' + str(temp.loc[0, 'MAIL_CITY']) + \
                                 ' ' + str(temp.loc[0, 'MAIL_STATE']) + ' ' + str(temp.loc[0, 'MAIL_ZIPCODE'])
        adds.loc[i, 'Use_Code'] = temp.loc[0, 'LU_DESC']
adds.head()     
```

```python
count = 0
for i in range(len(adds)):
    if adds.loc[i, 'Owner.1'] == adds.loc[i, 'Owner.2']:
        count += 1
    else:
        print(adds.loc[i, 'Owner.1'])
        print(adds.loc[i, 'Owner.2'])
        print('\n')
```

```python
count += 7
print(count/len(adds))
```

```python
count = 0
for i in range(len(adds)):
    if type(adds.loc[i, 'Owner.2']) == float and type(adds.loc[i, 'Address']) == float:
        count += 1

print(count/len(adds))
```

```python
pa_20 = pd.read_csv('Property Assesments/property-assessment-fy2020.csv')
pa_20 = pa_20.astype({"PID": str})
```

```python
for i in range(len(adds)):
    if type(adds.loc[i, 'Owner.2']) == float and type(adds.loc[i, 'Address']) == float:
        pid = adds.loc[i, 'parcel_id']
        filt = pa_20['PID'] == pid
        temp = pa_20[filt]
        temp = temp.reset_index(drop=False)
        if len(temp) == 0:
            pass
        else:
            adds.loc[i, 'Owner.2'] = temp.loc[0, 'OWNER']
            adds.loc[i, 'Address'] = str(temp.loc[0, 'MAIL_ADDRESS']) + str(temp.loc[0, 'MAIL CS']) + \
                                        str(temp.loc[0, 'MAIL_ZIPCODE'])
            use_code = temp.loc[0, 'LU']
            use_desc = pa_21[pa_21['LU'] == use_code].reset_index(drop=False).loc[0, 'LU_DESC']
            adds.loc[i, 'Use_Code'] = use_desc
```

```python
count = 0
for i in range(len(adds)):
    if type(adds.loc[i, 'Owner.2']) == float and type(adds.loc[i, 'Address']) == float:
        count += 1

print(count/len(adds))
```

```python
counts = pd.DataFrame(columns = ['Owners', 'Count'])
counts['Owners'] = adds['Owner.2'].unique()

owners = adds.groupby('Owner.2')

for i in range(len(counts)):
    owner = counts.loc[i, 'Owners']
    try:
        lst = owners.get_group(owner)
        counts.loc[i, 'Count'] = len(lst)
    except KeyError:
        pass

counts = counts.drop(4).reset_index(drop=False)
counts = counts.sort_values(by=['Count'], ascending=False).reset_index(drop=False)
counts = counts.drop(['level_0', 'index'], axis=1)
display(counts)
```
