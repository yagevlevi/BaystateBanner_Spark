```python
count = 0
for i in range(1,23):
    temp = pd.read_csv('2019 City Council Wards/ward' + str(i) + '-2019.csv')
    try:
        for n in temp.columns[1:-1]:
            if i < 10:
                if int(n) < 10:
                    wp = '0' + str(i) + '0' + str(n)
                else:
                    wp = '0' + str(i) + str(n)
            else:
                if int(n) < 10:
                    wp = str(i) + '0' + str(n)
                else:
                    wp = str(i) + str(n)
            entry = [wp]
            for x in range(12):
                entry.append(temp.loc[x, n])
            entry.append(0)
            entry.append(0)
            df.loc[count] = entry
            count += 1
    except ValueError:
        print(i)
        display(temp)
df = df.drop('Total', axis=1)
df = df.drop(254)
df
```
