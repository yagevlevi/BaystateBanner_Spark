```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('City Council/2019_November Municipal (General) City Council At Large Results.csv')
df = df[df.columns[0:17]]
cvap = pd.read_csv('2015-2019_Precinct_CVAP.csv')
df = df.drop('Unnamed: 0', axis=1)
#df = df.drop('Unnamed: 0.1', axis=1)
#df = df.drop('Unnamed: 0.1.1', axis=1)
df['CVAP'] = np.zeros(len(df))
df['TOTAL POP.'] = np.zeros(len(df))
df['WHITE'] = np.zeros(len(df))
df['AFRICAN_AMERICAN'] = np.zeros(len(df))
df['NATIVE_AMERICAN'] = np.zeros(len(df))
df['ASIAN'] = np.zeros(len(df))
df['OTHER'] = np.zeros(len(df))
df
```

```python
for i in range(len(cvap)):
    wp = cvap.loc[i, 'WARD_PRECI']
    if wp == '0502A':
        wp = '0502'
    filt = df['Ward/Precinct'] == int(wp)
    ind = df.index[filt].tolist()[0]
    df.loc[ind, 'CVAP'] += cvap.loc[i, 'CVAP']
```

```python
dems = pd.read_csv('Precinct_Demographics_Based_On_2015-19_ACS_Census_Data.csv')
dems = dems.drop('Unnamed: 0', axis=1)

for i in range(len(dems)):
    for c in dems.columns[2:]:
        dems.loc[i, c] = round(dems.loc[i, c]/dems.loc[i, 'TOTAL'], 3)

for i in range(len(dems)):
    wp = cvap.loc[i, 'WARD_PRECI']
    if wp == '0502A':
        wp = '0502'
    filt = df['Ward/Precinct'] == int(wp)
    ind = df.index[filt].tolist()[0]
    df.loc[ind, 'TOTAL POP.'] += dems.loc[i, 'TOTAL']
    df.loc[ind, 'WHITE'] += dems.loc[i, 'WHITE']
    df.loc[ind, 'AFRICAN_AMERICAN'] += dems.loc[i, 'AFRICAN_AMERICAN']
    df.loc[ind, 'NATIVE_AMERICAN'] += dems.loc[i, 'NATIVE_AMERICAN']
    df.loc[ind, 'ASIAN'] += dems.loc[i, 'ASIAN']
    df.loc[ind, 'OTHER'] += dems.loc[i, 'OTHER']
```

```python
df.to_csv('City Council/2019_November Municipal (General) City Council At Large Results.csv')
```
